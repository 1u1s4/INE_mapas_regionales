```python
import geopandas as gpd
import matplotlib.pyplot as plt

pth = 'regiones/regiones.shp'
tracts = gpd.read_file(pth)
valores = [10, 20, 30, 40, 50, 40, 12, 67]  # Crea una lista de valores que deseas agregar a la columna

tracts = tracts.assign(nueva_columna=valores)  # Agrega la columna de valores al GeoDataFrame

fig, ax = plt.subplots()  # Crea una figura y ejes

# Desactiva la leyenda y los bordes del mapa
tracts.plot(column="nueva_columna", cmap="OrRd", edgecolor="k", legend=False, ax=ax)
ax.axis('off')  # Desactiva los ejes
plt.show()  # Muestra la figura
```
